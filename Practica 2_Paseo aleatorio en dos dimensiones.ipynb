## **PASEO ALEATORIO EN DOS DIMENSIONES**
### Se parte del punto (1,1), cada paso que se da es de longitud r  

```python
import random  
import math  
import matplotlib.pyplot as plt  
random.seed()  
n=1000 # número de pasos  
r=10 # longitud del paso  
x=y=1 # punto de partida (x,y)=(1,1)  
listax=[x]  
listay=[y]  
for i in range(n):  
  angulo=random.random()*math.pi*2 
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  listax.append(x)  
  listay.append(y)  
plt.plot(listax,listay)  
plt.show()
```
